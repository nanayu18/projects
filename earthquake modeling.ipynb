```python
import matplotlib as plt
```

```python
cal = df[(df['location.name'] == 'California')]
cal.plot.scatter(x='location.longitude', y='location.latitude')
```

```python
import seaborn as sns
sns.kdeplot(
    x=cal["location.longitude"],
    y=df["location.latitude"],
    fill=True,
    cmap="hot",
    thresh=0
)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Heatmap of Earthquakes over California")
```

```python
sns.catplot(
    x=cal['time.day'],
    y=cal['impact.magnitude'])
sns.regplot(
    x=cal['time.day'],
    y=cal['impact.magnitude'],scatter=False, color = 'red')
```
